```python
import matplotlib.pyplot as plt
import numpy as np
```

Wykres zadania 1B

```python
def f(x):
    return np.sqrt(np.sqrt(-(x**2)+5*x-6))

xlist = np.linspace(2,3,num=15)
ylist = f(xlist)

plt.scatter(xlist,ylist, color='r',marker='o')
plt.title(r"f(x)=$\sqrt[4]{-x^{2}+5x-6}$")
plt.xlim(2,3)

plt.xlabel("Dziedzina")
plt.ylabel("Zbiór wartości")
plt.show()
```

Wykres zadania 2B

```python
def f(x):
    return x+1/x

xlist = np.linspace(-5,5,num=101)
ylist = f(xlist)

plt.plot(xlist,ylist,label=r'$f(x)=x+\frac{1}{x}$',color='g',linewidth=3)
plt.vlines(0,-11,11,color='r',linestyles='dotted',label='Asymptota x=0')
plt.ylim(-10,10)
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Wykres zadania 1D


```python
def f(x):
    return 1/np.sqrt((2**x)-(1/2))

xlist=np.linspace(0,20,100)
ylist=f(xlist)

plt.plot(xlist,ylist,color='m',label=r'$f(x)=\frac{1}{\sqrt{2^{x}-\frac{1}{2}}}$',linestyle='dashed')
plt.annotate('funkcja dąży do y=0',xy=(20,f(20)),xytext=(10, 0.4),
             arrowprops=dict(facecolor='red', shrink=0.05),)
plt.ylabel('oś rzędnych')
plt.xlabel('oś odcietych')
plt.title('Funkcja wymierna')
plt.legend()
plt.show()
```

Wykres zadania 3A

```python
def f(x):
    return np.power(x,2)
def g(x):
    return x+3

xlist=np.linspace(-20,20,1000)
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, dpi=160)

ax1.plot(xlist,f(xlist))
ax1.title.set_text('$f(x)=x^{2}$')

ax2.plot(xlist,g(xlist))
ax2.title.set_text('$g(x)=x+3$')

ax3.plot(xlist,f(f(xlist)), label = r'$f \circ f$')
ax3.plot(xlist,f(xlist),label=r'$f(x)$')
ax3.set_ylim(-10,100)
ax3.title.set_text(r'$f \circ f$')
ax3.legend()

ax4.plot(xlist,g(xlist),label=r'$g(x)$')
ax4.plot(xlist,g(g(xlist)),label=r'$g \circ g$')
ax4.title.set_text(r'$g \circ g$')
ax4.legend()

plt.subplots_adjust(wspace=0.5, hspace=0.5)


```

```python
def f(x):
    return np.sin(x)
def g(x):
    return np.exp(x+2)

xlist=np.linspace(-20,20,1000)

fig=plt.figure(facecolor='#808080',dpi=160)
plt.axvline(x=0,color='black',linewidth=0.6, linestyle='dashed')
plt.axhline(y=0,color='black',linewidth=0.6,linestyle='dashed')
plt.xlim(-6,6)

plt.plot(xlist,f(g(xlist)),label=r'$f \circ g$',color='green')
plt.plot(xlist,g(f(xlist)),label=r'$g \circ f$',color='red')
plt.title('Złożenia '+r'$f(x)= \sin(x)$' +" i "+ r'$g(x)=e^{x+2}$')
plt.legend()
plt.show()
```

Wykres zadania 5A

```python
def f(x):
    return np.cbrt(x)
def g(x):
    return np.log(x)
def s(x):
    return np.power(x,2)+5

fig=plt.figure(dpi=160)
xlist=np.linspace(-5,40,100)
plt.axvline(x=0,color='black',linewidth=0.6, linestyle='dashed')
plt.axhline(y=0,color='black',linewidth=0.6,linestyle='dashed')
plt.title('Przykład funkcji złożonej')

plt.plot(xlist,f(xlist),linestyle='dotted',label=r'$f(x)=\sqrt[3]{x}$')
plt.plot(xlist,g(xlist),linestyle='dashdot',label=r'$g(x)=\ln(x)$')
plt.plot(xlist,s(xlist),linestyle='dashed',label=r'$s(x)=x^{2}+5$')
plt.plot(xlist,g(s(xlist)),label=r'$g \circ s$')
plt.plot(xlist,f(g(s(xlist))),label=r'$f \circ (g \circ s)$',linewidth=2.6)

plt.ylim(-2,8)
plt.xlim(-5,40)
plt.legend()
plt.show()
```
