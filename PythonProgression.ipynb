Wizualizacja Ciągów z użyciem biblioteki matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import warnings
warnings.filterwarnings('ignore')
```

```python
nlist=np.linspace(1,100,num=100)
```

```python
def a(n):
    return (3-n)/(n+4)

mpl.rc('figure', figsize=(20, 5))
plt.scatter(nlist,a(nlist),color='red')
plt.title('Zadanie 4A - ciąg zbieżny do -1')
plt.show()
```

```python
def a(n):
    return np.sqrt(np.power(n,2)+4*n+1)-n

mpl.rc('figure', figsize=(20,5))
plt.scatter(nlist,a(nlist), color='green', marker='*')
plt.title('Zadanie 7D - ciąg zbieżny do 2')
plt.show()
```

```python
def a(n):
    return np.power(2,n)/np.power(n,3)
mpl.rc('figure', figsize=(20,5))
plt.scatter(nlist,a(nlist), color='orange', marker='h') 
plt.title(r'Zadanie 3B - ciąg rozbieżny do $+\infty$')
plt.show()
```
